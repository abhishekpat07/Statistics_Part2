#1. What is hypothesis testing in statistics?
- Hypothesis testing is a type of statisical analysis in which you put your assumptions about a population parameter to the test. It is used to estimate the relationship between two statistical variables.

#2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis of a test always predicts no effect or no relationship between variables, while the alternative hypothesis states your research prediction of an effect or relationship.

#3.  What is the significance level in hypothesis testing, and why is it important?
- The significance level is the probabilty of rejecting the null hypothesis when it the null hypothesis is true and is denoted by Î± .

#4. What does a P-value represent in hypothesis testing?
- A P-value is statistical measurment used to validate a hypothesis against observed data.

#5. How do you interpret the P-value in hypothesis testing?
- A P-value in hypothesis testing represents the probability of observing data as extreme as, or more extreme than, what was actually observed, assuming the null hypothesis is true.

#6. What are Type 1 and Type 2 errors in hypothesis testing?
- In Statistics, a Type 1 error is false positive conclusion, while a Type II error is a false negativev conclusion.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed tests allow for the possibility of an effect in one direction. Two-tailed test for the possibilty of an effect in two direction- positive and negative.

#8. What is the Z-test, and when is it used in hypothesis testing?
- A z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large.

#9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- A Z-score is calculated by subtracting the population mean from a data point and then dividing the result by the population standard deviation.

#10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The t-distribution describes the standardized distances of sample means to the population mean when the population standard deviation is not known, and the observations come from a normally distributed population. When working problems when the population standard deviation is not known and the sample size is small(n<30).

#11.  What is the difference between a Z-test and a T-test?
- A Z-test is used in statistics when the population standard deviation is known and typically applies to large sample sizes, while a T-test is used when the population standard deviation is unknown and is more suitable for small sample sizes.

#12. What is the T-test, and how is it used in hypothesis testing?
- A t-test is an inferential statistic used to determine if there is a significant difference between the means of two groups and how they are related.

#13. What is the relationship between Z-test and T-test in hypothesis testing?
- z-tests are used when we have large sample sizes (n > 30), whereas t-tests are most helpful with a smaller sample size (n < 30).

#14. What is a confidence interval, and how is it used to interpret statistical results?
- confidence interval shows the probability that a parameter will fall between a pair of values around the mean.If the confidence interval does not enclose the value reflecting 'no effect', this represents a difference that is statistically significant.

#15. What is the margin of error, and how does it affect the confidence interval?
- Margin of error is a statistical term that represents the range of uncertainty or variability around an estimate or measurement.

#16.  How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem thus gives the probability of an event based on new information that is or may be related to that event.

#17. What is the Chi-square distribution, and when is it used?
- A Chi-square distribution is a statistical distribution used primarily for hypothesis testing, particularly when analyzing categorical data, to determine if there is a significant difference between observed frequencies and expected frequencies in a dataset.

#18. What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness of fit test is a statistical hypothesis test used to determine whether a variable is likely to come from a specified distribution or not.

#19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is a continuous probability distribution used in statistics primarily to compare the variances of two populations.

#20. What is an ANOVA test, and what are its assumptions?

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):

    sample_size = len(sample)
    sample_mean = np.mean(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose from 'two-sided', 'greater', or 'less'.")

    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return {
        "Z-score": z_score,
        "p-value": p_value,
        "Decision": interpretation
    }

sample_data = [52, 55, 53, 58, 54, 56, 57, 50, 51, 53]
population_mean = 50
population_std = 5
alpha = 0.05
alternative = 'two-sided'

result = z_test(sample_data, population_mean, population_std, alpha, alternative)
print(result)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):

    sample_size = len(sample)
    sample_mean = np.mean(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose from 'two-sided', 'greater', or 'less'.")

    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return {
        "Z-score": z_score,
        "p-value": p_value,
        "Decision": interpretation
    }

def visualize_errors(alpha=0.05, beta=0.2):

    x = np.linspace(-4, 4, 1000)
    null_dist = norm.pdf(x, 0, 1)
    alt_dist = norm.pdf(x, 2, 1)

    critical_value = norm.ppf(1 - alpha)

    plt.figure(figsize=(10, 5))
    plt.plot(x, null_dist, label="Null Distribution (H0)", color='blue')
    plt.plot(x, alt_dist, label="Alternative Distribution (H1)", color='green')

    plt.fill_between(x, null_dist, where=(x > critical_value), color='red', alpha=0.3, label="Type 1 Error (Î±)")
    plt.fill_between(x, alt_dist, where=(x < critical_value), color='orange', alpha=0.3, label="Type 2 Error (Î²)")

    plt.axvline(critical_value, color='black', linestyle='dashed', label=f'Critical Value ({critical_value:.2f})')
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=5, size=30)
population_mean = 50
population_std = 5
alpha = 0.05
alternative = 'two-sided'

result = one_sample_z_test(sample_data, population_mean, population_std, alpha, alternative)
print(result)

visualize_errors(alpha=0.05, beta=0.2)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05, equal_var=True):

    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=equal_var)

    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return {
        "T-statistic": t_stat,
        "p-value": p_value,
        "Decision": interpretation
    }

def visualize_t_test(sample1, sample2):

    plt.figure(figsize=(10, 5))
    plt.hist(sample1, bins=15, alpha=0.5, label="Sample 1", color='blue', density=True)
    plt.hist(sample2, bins=15, alpha=0.5, label="Sample 2", color='green', density=True)
    plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', linewidth=2, label=f'Mean Sample 1 ({np.mean(sample1):.2f})')
    plt.axvline(np.mean(sample2), color='green', linestyle='dashed', linewidth=2, label=f'Mean Sample 2 ({np.mean(sample2):.2f})')
    plt.title("Distribution of Two Samples")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()


np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=53, scale=5, size=30)

alpha = 0.05

test_result = independent_t_test(sample1, sample2, alpha)
print(test_result)

visualize_t_test(sample1, sample2)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(before, after, alpha=0.05):

    t_stat, p_value = ttest_rel(before, after)

    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"

    return {
        "T-statistic": t_stat,
        "p-value": p_value,
        "Decision": interpretation
    }

def visualize_paired_t_test(before, after):

    plt.figure(figsize=(8, 5))
    plt.plot(before, after, 'bo', label="Paired Samples")
    plt.plot([min(before), max(before)], [min(before), max(before)], 'r--', label="No Change Line")
    plt.xlabel("Before")
    plt.ylabel("After")
    plt.title("Paired Sample Comparison")
    plt.legend()
    plt.show()

np.random.seed(42)
before = np.random.normal(loc=50, scale=5, size=30)
after = before + np.random.normal(loc=2, scale=3, size=30)
alpha = 0.05

test_result = paired_t_test(before, after, alpha)
print(test_result)

visualize_paired_t_test(before, after)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, ttest_rel, norm, t

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test to compare the sample mean with the population mean.
    """
    sample_size = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return {"Z-score": z_score, "p-value": p_value, "Decision": interpretation}

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent T-test to compare the means of two samples.
    """
    t_stat, p_value = ttest_ind(sample1, sample2)
    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return {"T-statistic": t_stat, "p-value": p_value, "Decision": interpretation}

def paired_t_test(before, after, alpha=0.05):
    """
    Perform a paired sample T-test to compare two related samples.
    """
    t_stat, p_value = ttest_rel(before, after)
    interpretation = "Reject the null hypothesis" if p_value < alpha else "Fail to reject the null hypothesis"
    return {"T-statistic": t_stat, "p-value": p_value, "Decision": interpretation}

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.
    """
    sample_size = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    t_critical = t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    return (sample_mean - margin_of_error, sample_mean + margin_of_error)

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.
    """
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)
    t_critical = t.ppf((1 + confidence) / 2, df=sample_size - 1)
    return t_critical * (sample_std / np.sqrt(sample_size))

def visualize_distributions(sample1, sample2):
    """
    Visualizes the distribution of two samples.
    """
    plt.figure(figsize=(10, 5))
    plt.hist(sample1, bins=15, alpha=0.5, label="Sample 1", density=True, color='blue')
    plt.hist(sample2, bins=15, alpha=0.5, label="Sample 2", density=True, color='green')
    plt.axvline(np.mean(sample1), color='blue', linestyle='dashed', label=f'Mean Sample 1 ({np.mean(sample1):.2f})')
    plt.axvline(np.mean(sample2), color='green', linestyle='dashed', label=f'Mean Sample 2 ({np.mean(sample2):.2f})')
    plt.title("Comparison of Sample Distributions")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# Simulate random sample data
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=53, scale=5, size=30)
paired_before = np.random.normal(loc=50, scale=5, size=30)
paired_after = paired_before + np.random.normal(loc=2, scale=3, size=30)

alpha = 0.05
population_mean = 50
population_std = 5

# Perform Z-test
z_test_result = one_sample_z_test(sample1, population_mean, population_std, alpha)
print("One-Sample Z-test Result:", z_test_result)

# Perform independent T-test
t_test_result = independent_t_test(sample1, sample2, alpha)
print("Independent T-test Result:", t_test_result)

# Perform paired T-test
paired_test_result = paired_t_test(paired_before, paired_after, alpha)
print("Paired Sample T-test Result:", paired_test_result)

# Calculate confidence interval
ci = confidence_interval(sample1)
print(f"Confidence Interval for Sample 1 Mean: {ci}")

# Calculate margin of error
moe = margin_of_error(sample1)
print(f"Margin of Error for Sample 1 Mean: {moe:.4f}")

# Visualize distributions
visualize_distributions(sample1, sample2)

```

```python
#16.  Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=10, size=100)
sample2 = np.random.normal(loc=50, scale=20, size=100)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F_statistic = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(sample1) - 1
df2 = len(sample2) - 1

p_value = 1 - f.cdf(F_statistic, df1, df2)

print(f"Variance of Sample 1: {var1}")
print(f"Variance of Sample 2: {var2}")
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("The variances of the two samples are significantly different.")
else:
    print("\nSince the p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant difference between the variances of the two samples.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene, probplot

def check_anova_assumptions(groups):
    print("Checking Normality (Shapiro-Wilk test):")
    for i, group in enumerate(groups):
        stat, p_value = shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk Test p-value: {p_value:.4f}")
        if p_value < 0.05:
            print(f"  Group {i+1} does NOT follow a normal distribution (p < 0.05).\n")
        else:
            print(f"  Group {i+1} follows a normal distribution (p >= 0.05).\n")

    print("Checking Equal Variance (Levene's test):")
    stat, p_value = levene(*groups)
    print(f"Levene's test p-value: {p_value:.4f}")
    if p_value < 0.05:
        print("  The variances are significantly different (p < 0.05).\n")
    else:
        print("  The variances are equal (p >= 0.05).\n")

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title('Boxplot of Groups')
    plt.xlabel('Groups')
    plt.ylabel('Values')

    plt.subplot(1, 2, 2)
    for i, group in enumerate(groups):
        probplot(group, dist="norm", plot=plt)
        plt.title(f'Q-Q Plot of Group {i+1}')
        plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=12, size=30)
group3 = np.random.normal(loc=60, scale=15, size=30)

check_anova_assumptions([group1, group2, group3])
```

```python
#20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f

import matplotlib.pyplot as plt
import seaborn as sns

# Simulating data for two-way ANOVA
np.random.seed(42)

# Factor A (2 levels)
factor_a = ['A1', 'A2'] * 30  # 2 levels, 30 samples per level

# Factor B (3 levels)
factor_b = ['B1', 'B2', 'B3'] * 20  # 3 levels, 20 samples per level

# Dependent variable (DV)
# Random values from normal distribution for each combination of factor levels
dependent_variable = np.random.normal(loc=50, scale=10, size=60)

# Create a DataFrame
data = pd.DataFrame({
    'Factor_A': factor_a,
    'Factor_B': factor_b,
    'Dependent_Variable': dependent_variable
})

# Perform Two-Way ANOVA
model = ols('Dependent_Variable ~ Factor_A * Factor_B', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)  # Type II ANOVA table
print(anova_table)

# Visualize the interaction using an interaction plot
plt.figure(figsize=(10, 6))
sns.interactionplot(x=data['Factor_A'], y=data['Dependent_Variable'],
                    hue=data['Factor_B'], markers=['o', 's', 'D'], linestyles=['-', '--', ':'])
plt.title('Interaction Plot between Factor A and Factor B')
plt.xlabel('Factor A')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the resultsD.

import numpy as np
import scipy.stats as stats

np.random.seed(42)

sample_data = np.random.normal(loc=50, scale=10, size=30)

sigma_squared_0 = 100

sample_variance = np.var(sample_data, ddof=1)

n = len(sample_data)

chi_square_stat = (n - 1) * sample_variance / sigma_squared_0

df = n - 1

p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample variance and the hypothesized variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import scipy.stats as stats

x1 = 80
n1 = 100

x2 = 90
n2 = 120

p1 = x1 / n1
p2 = x2 / n2

P = (x1 + x2) / (n1 + n2)

z_stat = (p1 - p2) / np.sqrt(P * (1 - P) * (1 / n1 + 1 / n2))

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```
